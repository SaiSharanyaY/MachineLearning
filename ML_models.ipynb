## Overfitting

1. happens when model fits the data too well, it learns the detail and noise in the training dataset to the extent that it negatively impacts the performance of the model.
2. training data accurary >> test data accurarcy


Causes for overfitting

1. less data
2. increased complexity of the model
3. more number of layers in neural network

preventing overfitting by 

1. using more data
2. reduce no. of layers in neural network
3. early stopping
4. high variance low bias
5. using dropouts - used in neural networks


## **Underfitting**

happens when model does not learn enough from data, occurs when the model cannot capture underlying trend in data.



Causes for underfitting :

1. choosing wrong model
2. less complexity of model
3. less variance but high bias


prevent underfitting by

1. choosing correct model for problem
2. increasing the complexity of the model
3. more no. of parameters of model.
4. bias variance trade-off

## BIAS VARIANCE TRADE OFF

**Bias** 
it is the diff btwn avg prediction of our model and correct value which we are trying to predict.

**variance**
it is the amount that the estimate of the target function will change if different training data was used.

Techniques to have better bias-variance trade off

1. good model selction.
2. regularization.
3. dimensionality reduction.
4. ensemble methods - combining 2/3 models.

## Gradient Descent

- optimisation refers to determining best parameters for a model, such that the loss functn of the model decreases, as a result model can predict more accurately.

gradient descent is an optimization algorithm used for minimizing the loss function in various machine learning algorithms. it is used for updating the parameters of the learning model.

- w = w - L*dw
- b = b - L*db

- dw - partial derivative of loss function with respect to m.
- db - partial derivative of loass function with respect to c.
